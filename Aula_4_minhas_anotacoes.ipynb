```python
# Elimina os números e as pontuações (exceto underline)
from nltk.tokenize import RegexpTokenizer
texto='Com um passe de Eli Manning para Plaxico Burress a 39 segundos do fim, o New York Giants anotou o touchdown decisivo e derrubou o favorito New England Patriots por 17 a 14 neste domingo, em Glendale, no Super Bowl XLII. O resultado, uma das maiores zebras da história do Super Bowl, acabou com a temporada perfeita de Tom Brady e companhia, que esperavam fazer história ao levantar o troféu da NFL sem sofrer uma derrota no ano. '
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+')
# tokenizer = RegexpTokenizer(r'[A-z]\w+')  # idêntico
tokens = tokenizer.tokenize(texto)
print(tokens)
```

```python
# Frequência / contagem
frequencia = nltk.FreqDist(tokens)
print(frequencia.most_common())
```

```python
corpus = open('corpus_teste.txt').read()
print(corpus)
```

```python
# Elimina as pontuações (exceto underline)
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(corpus)

frequencia = nltk.FreqDist(tokens)
print(frequencia.most_common())
```

```python
# caixa baixa = minúsculas
# Elimina os números e as pontuações (exceto underline)
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+')
tokens = tokenizer.tokenize(corpus)

nova_lista = []
for token in tokens:
    nova_lista.append(token.lower())

frequencia = nltk.FreqDist(nova_lista)
print(frequencia.most_common())
```

```python
# caixa baixa = minúsculas
# Elimina os números e as pontuações (exceto underline)
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+')
tokens = tokenizer.tokenize(corpus)
stopwords = nltk.corpus.stopwords.words('portuguese')

nova_lista = []
for token in tokens:
    if token.lower() not in stopwords:
        nova_lista.append(token.lower())
        
frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common()
```

```python
# com list comprehension
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w+')
tokens = tokenizer.tokenize(corpus)
stopwords = nltk.corpus.stopwords.words('portuguese')

nova_lista = []

# for token in tokens:
#     if token.lower() not in stopwords:
#         nova_lista.append(token.lower())
tokens_sem_stopwords = (w.lower() for w in tokens if w.lower() not in stopwords)
        
# frequencia = nltk.FreqDist(nova_lista)
frequencia = nltk.FreqDist(tokens_sem_stopwords)
frequencia.most_common()
```

```python
# N-grams: palavras que ocorrem conjuntamente
# bigrams, trigrams, ngrams (4+)
from nltk import bigrams
from nltk import trigrams
from nltk import ngrams
list(bigrams(tokens))
list(trigrams(tokens))
list(ngrams(tokens,4))
```
