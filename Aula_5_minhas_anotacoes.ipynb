**Etiquetadores**

Palavra e classe gramatical.

```python
from nltk.corpus import mac_morpho
```

```python
from nltk.tag import UnigramTagger
```

```python
from nltk.tag import DefaultTagger
```

```python

corpus = open('corpus_teste.txt', 'r', encoding='utf-8')
```

```python
tokens = nltk.word_tokenize(corpus.read())
```

```python
etiq_padrao = DefaultTagger('N')
sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino, backoff=etiq_padrao)

etiquetado=etiquetador.tag(tokens)

print(etiquetado)
```

```python
# Retorna somente números
number_tokens = [token.orth_ for token in doc if token.is_digit]
print(number_tokens)
```

```python
# Retorna somente pontuações
punct_tokens = [token.orth_ for token in doc if token.is_punct]
print(punct_tokens)
```

```python
corpus = open('corpus_teste.txt',  encoding='utf-8').read()
corpus
```

```python
nlp = spacy.load('pt_core_news_sm')
doc = nlp(corpus)
```

```python
# Retorna strings
tokens = [token.orth_ for token in doc]
print(tokens)
```

```python
# Retorna somente letras/palavras
alpha_tokens = [token.orth_ for token in doc if token.is_alpha]
print(alpha_tokens)
# Retorna somente números
number_tokens = [token.orth_ for token in doc if token.is_digit]
print(number_tokens)
# Retorna somente pontuações
punct_tokens = [token.orth_ for token in doc if token.is_punct]
print(punct_tokens)
```

**Lemmatização**

Só funciona para os verbos

Spacy não tem Stemming

```python
lemmas = [token.lemma_ for token in doc if token.pos_ == 'VERB']
print(lemmas)
```

**Etiquetas**

Criando tuplas com as Strings e as Classes Gramaticais (pos = part of speech)

```python
etiquetas = [(token.orth_, token.pos_) for token in doc]
print(etiquetas)
```

**Classificações morfológicas**

```python
etiquetas = [(token.orth_, token.morph) for token in doc]
print(etiquetas)
```

**Entidades nomeadas = Chunk do NLTK**

```python
entidades = list(doc.ents)
print(entidades)
```

```python
detalhes = [(entidade, entidade.label_) for entidade in doc.ents]
print(detalhes)
```

**Visualizar entidades nomeadas**

Parser é uma ferramenta que realiza a análise do texto no nível da sentença. Esse tipo de análise é chamada de análise sintática.<br>

A sintaxe analisada pode ser de dois tipos: dependência ou constituintes. A análise por dependência encontra as relações entre as palavras em termos de função sintática (sujeito, objeto direto, objeto indireto, aposto, predicativo, complemento nominal, etc...). Já a análise por constituintes encontra as a relação das palavras em termos de sintagmas, nominais, verbais, preposicionados ou não, etc...<br>

```python
from pathlib import Path
```

```python
html = spacy.displacy.render(doc, style='ent')
output_path = Path('entidades_nomeadas.html')
output_path.open('w', encoding='utf-8').write(html)
```

**Análise sintática**

```python
sintatica = [(token.orth_, token.dep_) for token in doc]
print(sintatica)
```

```python
svg = spacy.displacy.render(doc, style='dep')
output_path = Path('sintatica.svg')
output_path.open('w', encoding='utf-8').write(svg)
```
